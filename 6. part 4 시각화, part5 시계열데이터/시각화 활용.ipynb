```python
# 사인 함수 그리기

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

import seaborn as sns

x = np.linspace(0, 14, 100)
y1 = np.sin(x)
y2 = 2*np.sin(x+0.5)
y3 = 3*np.sin(x+1.0)
y4 = 4*np.sin(x+1.5)

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
sns.set_style("dark")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)

sns.despine() # 필요없는 axis border(top, right 두 축) 제거

plt.show()
```

```python
sns.set_style("whitegrid")

plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)
plt.show()
```

```python
plt.figure(figsize=(10,6))
plt.plot(x,y1, x,y2, x,y3, x,y4)

sns.despine(offset=10) # x,y 축을 살짝 벌린다.

plt.show()
```

```python
tips = sns.load_dataset("tips")
tips.head(5)
```

```python
# boxplot 그리기
sns.set_style("whitegrid")

plt.figure(figsize=(8,6))
sns.boxplot(x=tips["total_bill"])
plt.show()
```

```python
# x축은 요일로, y축은 전체 금액으로 설정
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", 
            hue="smoker", # hue 옵션으로 흡연 여부로 구분
            data=tips, 
            palette="Set3") # palette는 색상
plt.show()
```

```python
# swarmplot 그리기
plt.figure(figsize=(8,6))
sns.swarmplot(x="day", y="total_bill", data=tips, color=".5")
plt.show()
```

```python
# swarmplot, boxplot 동시에 그리기
plt.figure(figsize=(8,6))
sns.boxplot(x="day", y="total_bill", data=tips)
sns.swarmplot(x="day", y="total_bill", data=tips, color=".25")
plt.show()
```

```python
# implot 그리기
sns.set_style("darkgrid")
sns.lmplot(x="total_bill", y="tip", data=tips)
plt.show()
```

```python
sns.lmplot(x="total_bill", y="tip", hue="smoker", data=tips, palette="Set1")
plt.show()
```

```python
flights = sns.load_dataset("flights") # 연도 및 월별 항공기 승객 수를 기록한 데이터
flights.head(5)
```

```python
# pivot하여 월별, 연도별로 구분
flights = flights.pivot("month", "year", "passengers")
flights.head(5)
```

```python
plt.figure(figsize=(10,8))
sns.heatmap(flights, annot=True, fmt="d") # annot:각 셀에 숫자 입력,  fmt='d' 는 정수 형태
plt.show()
```

```python
iris = sns.load_dataset("iris")
iris.head(10)
```

```python
sns.set(style='ticks')
sns.pairplot(iris, hue="species")
plt.show()
```

```python
sns.pairplot(iris, vars=["sepal_width", "sepal_length"]) # 몇 가지 변수만 적용하기
plt.show()
```

```python
# 변수 구체적으로 적용하기
sns.pairplot(iris, 
             x_vars=["sepal_width", "sepal_length"], 
             y_vars=["petal_width", "petal_length"])
plt.show()
```

```python
anscombe = sns.load_dataset("anscombe")
anscombe.head(5)
```

```python
sns.set_style("darkgrid")
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'I'"),
           ci=None)
plt.show()
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'II'"),
           order=2, # 2차함수로 표현
           ci=None)
plt.show()
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           scatter_kws={'s':80},
          ci=None) # 크기 조절
plt.show()
```

```python
sns.lmplot(x="x", y="y", data=anscombe.query("dataset == 'III'"),
           scatter_kws={'s':80},
           robust = True, # robust 옵션을 이용해서 너무 튀는 데이터는 무시
          ci = None) 
plt.show()
```
